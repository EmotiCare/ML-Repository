```python
import os
import json

# Daftar file yang diunggah
dataset1_path = '/kaggle/input/dataset/data1.json'
dataset2_path = '/kaggle/input/dataset/data2.json'
dataset3_path = '/kaggle/input/dataset/data3.json'

# Membaca dataset dari file JSON
with open(dataset1_path, 'r') as file:
    data1 = json.load(file)

with open(dataset2_path, 'r') as file:
    data2 = json.load(file)

with open(dataset3_path, 'r') as file:
    data3 = json.load(file)
    
# Menggabungkan intents dari tiga dataset
combined_intents = data1['intents'] + data2['intents'] + data3['intents']

# Fungsi untuk menghapus duplikasi
def remove_duplicates(intents):
    seen = set()
    unique_intents = []
    for intent in intents:
        patterns_tuple = tuple(intent['patterns'])
        if patterns_tuple not in seen:
            seen.add(patterns_tuple)
            unique_intents.append(intent)
    return unique_intents

# Menghapus duplikasi
combined_intents = remove_duplicates(combined_intents)

# Membuat dataset gabungan
combined_dataset = {"intents": combined_intents}

# Menyimpan dataset gabungan ke file JSON baru
output_path = '/kaggle/working/combined_dataset.json'
with open(output_path, 'w') as file:
    json.dump(combined_dataset, file, indent=4)

print(f"Dataset gabungan disimpan di {output_path}")
```

```python
import pandas as pd
import json

# Memuat dataset JSON dari file
dataset = '/kaggle/input/dataset-akhir/intense.json'
with open(dataset, 'r') as file:
    data = json.load(file)

# Ekstraksi intents dari data JSON
intents = data['intents']

# Membuat list kosong untuk menampung data yang diekstraksi
data_extracted = []

# Iterasi melalui setiap intent dan ekstraksi tag, patterns, dan responses
for intent in intents:
    tag = intent['tag']
    patterns = intent['patterns']
    responses = intent['responses']
    
    # Mengorganisir data untuk DataFrame
    for pattern in patterns:
        data_extracted.append({
            'Tag': tag,
            'Patterns': pattern,
            'Responses': ', '.join(responses) 
        })

# Membuat DataFrame dari data yang diekstraksi
df = pd.DataFrame(data_extracted)

# Menampilkan DataFrame dalam bentuk tabel yang rapi
print("\nSample of the Dataset:")
display(df.sample(10))  # Menampilkan 10 baris sampel dari DataFrame

# Menampilkan informasi shape dari DataFrame
print("\nShape of the DataFrame:", df.shape)

# Menampilkan jumlah total baris dalam DataFrame
print("\nTotal number of rows in the Dataset:", df.shape[0])
```
